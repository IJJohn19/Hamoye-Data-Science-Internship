```python
#importing relevant libraries
import numpy as np #for array numerical computing and array operations
import pandas as pd #for two-dimensional arrays and tabular operations
import matplotlib.pyplot as plt #scientific operations and plotting
import seaborn as sns #data visualisation
import sklearn
print('relevant libraries imported')
```

```python
#importing the quiz dataset
df_main = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00374/energydata_complete.csv')
df_main.head()
```

```python
#Renaming some columns
df_main = df_main.rename(columns = {'T1':'kitchen_temp', 'RH_1':'kitchen_hum', 'T2':'livrm_temp', 'RH_2':'livrm_hum', 'T3': 'laundry_temp',
                         'RH_3':'laundry_hum', 'T4':'office_temp', 'RH_4':'office_hum', 'T5': 'bath_temp', 'RH_5': 'bath_hum',
                         'T6': 'outside_temp', 'RH_6':'outside_hum', 'T7':'ironrm_temp', 'RH_7': 'ironrm_hum', 'T8':'teenrm_temp',
                         'RH_8':'teenrm_hum', 'T9':'dadsrm_temp','RH_9':'dadsrm_hum', 'T_out':'Chis_temp', 'RH_out':'Chis_hum' })
print('First 5 rows of main dataset:')
df_main.head()
```

```python
df_main.describe() #to get basic statistical info about the dataset
```

```python
df_main.shape
```

```python
#removing recommended columns
df = df_main.drop(['date', 'lights'], axis = 1)
df.head()
```

```python
#performing recommended scaling of the dataset with the MinMax Scaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler() #to instantiate an object of the scaler
df = pd.DataFrame(scaler.fit_transform(df), columns = df.columns)
print('First 5 rows of scaled dataset:')
df.head()
```

```python
df.shape
```

Notice that the number of columns has reduced by 2 after dropping date and light columns.

```python
#specifying target and response variables
X = df.drop('Appliances', axis = 1) #predictors
X.head()
```

```python
#specifying response variable
y = df['Appliances']
y.head()
```

```python
#splitting the dataset, given test size of 30% and random_state parameter value of 42
from sklearn.model_selection import train_test_split as tts
X_train, X_test, y_train, y_test = tts(X,y, test_size = 0.3, random_state = 42 ) 
```

```python
#fitting the training dataset to a linear regression model
from sklearn.linear_model import LinearRegression
mlr = LinearRegression()
mlr.fit(X_train,y_train)
```

```python
#trying the model on the test set
y_pred = mlr.predict(X_test) #to predict y_test
y_pred
```

```python
#evaluating the model on the test set
mlr_score = mlr.score(X_test,y_test) #the score of the model on the test set is the r-squarred value
print ('R-squared value of Multiple Linear Regression or OLS: {}'.format(round(mlr_score,2)))
#alternative method for r-square and other evaluation metrics 
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import mean_squared_error as msr
print('Confirming, R-squared value of Ordinary Least Square (OLS) Regression  is: {}'.format(round(r2_score(y_test, y_pred),3)))
print('OLS mean absolute error: {:.2f}'.format(mae(y_test, y_pred)))
print ('OLS mean squared error: {:.2f}'.format(msr(y_test, y_pred)))
print('OLS sum of squared residuals: {:.2f}'.format(np.sum(np.square(y_test-y_pred))))
from sklearn.metrics import  mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print('rmse_linreg:', round(rmse),3) #this error cost me two correct answers:inputing 0.03 instead of 0.088 and selecting 'yes' in another question that the ridge rmse differs from the linear reg rmse
```

The above R-squarred value obviously show low performance of linear regression or ordinary least square regression on the dataset.

```python
#this function creates a table of the weight of every feature in the ols model
def get_weights_df(model,feet,col_name):
    weights = pd.Series(model.coef_,feet.columns).sort_values() #creates a pandas series of the coefficients of the model and columns of the feet argument
    weights_df = pd.DataFrame(weights).reset_index() #creates a dataframe from the weigths and makes adds an index column 
    weights_df.columns = ['Features', col_name] #renames the columns headers to 'Features' and the column name
    weights_df[col_name].round(3) #rounds them to 3decimal places
    return(weights_df)

#calling the function for the three linear regression models
linear_regression_weights = get_weights_df(mlr, X_train, 'Linear Model Weight') #calls the function for the linear regression model


linear_regression_weights.sort_values('Linear Model Weight', ascending = True).head()
```
